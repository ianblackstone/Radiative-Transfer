I wanted to make sure we were picking the correct random distribution by taking $\mu$ uniformly between -1 and 1.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Determine the number of points to compare.  taking this too high makes the later linear plots too dense to read.
num = 250

# Generate random values in the distribution
ran_mu = 2*np.random.rand(num)-1
ran_theta = np.pi*np.random.rand(num)

# Need the angles directly in the mu diostribution.
ran_mu_theta = np.arccos(ran_mu)
```

```python
def PlotLines(ran):
    # Plot each line from [0,0] to the endpoint.
    for theta in ran:
        x = [0,np.cos(theta).tolist()]
        y = [0,np.sin(theta).tolist()]
        plt.plot(x,y,color='green')
        plt.xlim(-1,1)
        plt.ylim(0,1)
```

```python
PlotLines(ran_theta)
plt.title('uniform distribution in theta')
plt.show()
```

```python
PlotLines(ran_mu_theta)
plt.title('uniform distribution in mu')
plt.show()
```

This looks less uniformly distributed than the $\theta$ plot, it has clear gaps at high angles.

I want to repeat this with the randomness taken out of the equation, and look directly at the two distributions.

```python
# generate linearly spaced for comparison
ran_mu_lin = np.linspace(-1,1,num)
ran_theta_lin = np.linspace(0,np.pi,num)

# Need the angles directly in the mu diostribution.
ran_mu_theta_lin = np.arccos(ran_mu_lin)
```

```python
PlotLines(ran_theta_lin)
plt.title('uniform distribution in theta')
plt.show()
```

```python
PlotLines(ran_mu_theta_lin)
plt.title('uniform distribution in mu')
plt.show()
```

The uniform $\mu$ distribution clearly weights against forward and backward moving photons.  I also want top plot these as histograms so we can see the angular distribution.

```python
plt.hist(ran_theta,bins='auto')
plt.title('random values of theta')
plt.ylabel('count')
plt.xlabel('angle in radians')
plt.xticks([0,np.pi/4,np.pi/2,3*np.pi/4,np.pi],labels=['0',r"$\frac{\pi}{4}$",r'$\frac{\pi}{2}$',r'$\frac{3\pi}{4}$',r'$\pi$'])
```

```python
plt.hist(ran_mu_theta,bins='auto')
plt.title('random values of mu')
plt.ylabel('count')
plt.xlabel('angle in radians')
plt.xticks([0,np.pi/4,np.pi/2,3*np.pi/4,np.pi],labels=['0',r"$\frac{\pi}{4}$",r'$\frac{\pi}{2}$',r'$\frac{3\pi}{4}$',r'$\pi$'])
```
